Implement AND and OR logic operations using a single perceptron, and verify the correctness of the output using appropriate truth tables. (linear Data)

```python
def step(x):
  return 1 if x>=0 else 0


def And_percep(x1,x2):
  w1=1
  w2=1
  bias=-1.5

  com=x1*w1+x2*w2+bias
  return step(com)

print("AND GATE")
for x1 in [0,1]:
  for x2 in [0,1]:
    print(x1,x2,"->",And_percep(x1,x2))



def OR_perceptron(x1, x2):
    w1 = 1
    w2 = 1
    bias = -0.5

    net = x1*w1 + x2*w2 + bias
    return step(net)

print("\nOR Gate Output")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(x1, x2, "->", OR_perceptron(x1, x2))




```

Implement the XOR and (¬XOR) logic operation using a multi-perceptron network, and analyze how multiple perceptrons overcome the limitations of a single perceptron.


```python
def step(x):
    return 1 if x >= 0 else 0
def XOR(x1, x2):

    h1 = step(x1 + x2 - 0.5)
    h2 = step(x1 + x2 - 1.5)
    output = step(h1 - 2*h2 - 0.5)
    return output

print("XOR Output")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(x1, x2, "->", XOR(x1, x2))


def XNOR(x1, x2):
    return 1 - XOR(x1, x2)

print("\nXNOR Output")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(x1, x2, "->", XNOR(x1, x2))



```
